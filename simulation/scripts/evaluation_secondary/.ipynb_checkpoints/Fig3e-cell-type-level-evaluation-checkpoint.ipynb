```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math 
import os
from scipy.stats import spearmanr, pearsonr
from scipy.stats import entropy
```

```python
cell_types=['Cancer', 'Ductal', 'mDCs', 'Monocytes', 'Mast', 'Macrophages']
#tru_fn = "/mnt/humble_50t/alexw/ReSort_manuscript/simulation/simulation_finer_resolution/data/cancer_inf_0.1/simulated_mixture_cell_ground_truth.csv"
tru_fn = '../../simulated_data/infiltrated/simulated_mixture_cell_ground_truth_secondary.csv'
truth = pd.read_csv(tru_fn, index_col=0)
truth = truth[['Cancer', 'Ductal', 'count_Macrophages', 'count_Mast', 'count_mDCs', 'count_Monocytes']]
truth.columns = ['Cancer', 'Ductal', 'Macrophages', 'Mast', 'mDCs', 'Monocytes']
truth = truth.divide(truth.sum(axis=1), axis=0)
```

```python
def read_process_res(fp, method,
                     cell_types=['Cancer', 'Ductal', 'mDCs', 'Monocytes', 'Mast', 'Macrophages']):
    
    if method == 'stsc':
        sep = "\t"
    else:
        sep = ","
        
    df = pd.read_csv(fp, sep=sep, index_col=0)

    if method == 'spatialDWLS':
        df.set_index('cell_ID', drop=True, inplace=True)
        df.index.name = None
        
    df = df[cell_types]
    
    for i in range(df.shape[0]):
        if (df.iloc[i, :] == 0).all():
            df.iloc[i, :] = 0.1
            
    df = df.divide(df.sum(axis=1), axis=0)
    return df

def read_2steps(f, method, mode='corrected', fmt='csv'):
    
    sep = ","
    if fmt == 'tsv':
        sep = "\t"
        
    df_mist = pd.read_csv(f"{f}/props_mist.{fmt}", sep=sep, index_col=0)

    if mode == 'corrected':
        df2 = pd.read_csv(f"{f}/props_immune_corrected.{fmt}",sep=sep, index_col=0)
    else:
        df2 = pd.read_csv(f"{f}/props_immune.{fmt}",sep=sep, index_col=0)
        
    if method == 'spatialDWLS':
        df_mist.set_index("cell_ID", drop=True, inplace=True)
        df2.set_index("cell_ID", drop=True, inplace=True)
        
    df2 = df2[['mDCs', 'Monocytes', 'Mast', 'Macrophages']]
    df2 = df2.divide(df2.sum(axis=1), axis=0)

    df_mist = df_mist[['Cancer', 'Ductal', 'Stroma']]
    df_mist = df_mist.divide(df_mist.sum(axis=1), axis=0)
    
    stroma = df_mist['Stroma'].to_numpy()
    
    for c in df2.columns:
        df2[c] = df2[c] * stroma
    
    df2 = pd.concat([df_mist[['Cancer', 'Ductal']], df2], axis=1)
    df2.index.name=None
    df2 = df2.fillna(0.)
    return df2

def calculate_ct_errors(est, tru, 
                      cell_types=['Cancer', 'Ductal', 'mDCs', 'Monocytes', 'Mast', 'Macrophages']):
    """calculate cell type level Mean Absolute Percentage Error"""
    est = est.loc[tru.index, cell_types]
    tru = tru.loc[:, cell_types]
    mapes, maes, rs, kls = [], [], [], []
    
    for ct in cell_types:
        
        mp = np.mean(tru[ct].to_numpy())
        ae = np.absolute(tru[ct].to_numpy() - est[ct].to_numpy())
        mae = np.nanmean(ae) * 100
        mape = np.nanmean(ae/mp) * 100
        r = pearsonr(np.ravel(tru[ct].to_numpy()), np.ravel(est[ct].to_numpy()))[0]
        
        maes.append(mae)
        mapes.append(mape)
        rs.append(r)
        
        kl = entropy(tru[ct].to_numpy() + 0.001, est[ct].to_numpy() + 0.001)
        kls.append(kl)
    res_df = pd.DataFrame({'cell type': cell_types,
                            'MAPE': mapes,
                           'MAE': maes,
                           "r": rs,
                          "KL": kls},
                           index=cell_types)
    return res_df

```

```python
def ax_plot_radar(res, label, ax, col='MAPE', color='green'):
    theta = np.arange(len(res) + 1) / float(len(res)) * 2 * np.pi
    values = res[col].values
    values = np.append(values, values[0])
    l1, = ax.plot(theta, values, color=color, marker="o", label=label)
    ax.tick_params(pad=10)
    ax.fill(theta, values, color, alpha=0.3)
    
def get_result_method(method, plot=True, col='r'):
    fmt = 'tsv' if method == 'stsc' else 'csv'
    folder = f'/mnt/humble_50t/alexw/ReSort_manuscript/simulation/simulation_finer_resolution/results/{method}'
    ## internal
    int_df = read_process_res(f"{folder}/props_internal.{fmt}", method)
    int_res = calculate_ct_errors(int_df, truth)
    int_res['reference'] = 'Internal'
    ## external
    ext_df = read_process_res(f"{folder}/props_external.{fmt}", method)
    ext_res = calculate_ct_errors(ext_df, truth)
    ext_res['reference'] = 'External'
    ## 2-step
    step2_df = read_2steps(folder, method, '', fmt)
    step2_res = calculate_ct_errors(step2_df, truth)
    step2_res['reference'] = 'ReSort'
    
    res_dfs = pd.concat([int_res, ext_res, step2_res])
    res_dfs['method'] = method
    
    if plot:
        plt.rcParams["figure.figsize"] = [7.50, 3.50]
        plt.rcParams["figure.autolayout"] = True
        fig = plt.figure(figsize=(10, 10))
        ax = fig.add_subplot(111, projection="polar")
        ax_plot_radar(int_res, 'Internal', ax, col, color='red')
        ax_plot_radar(ext_res, 'External', ax, col, color='orange')
        ax_plot_radar(step2_res, 'ReSort', ax, col, color='green')
        angles = [n / float(len(int_res)) * 2 * math.pi for n in range(len(int_res))]
        plt.xticks(angles, int_res.index.tolist(), fontsize=15)
        plt.legend(bbox_to_anchor=(1.4, 0.8), frameon=True, fontsize=15)
        return res_dfs, fig
    return res_dfs
```

```python
res_df,  f = get_result_method('stsc', True, col='r')
res_df = res_df[['cell type', 'r', 'KL', 'reference']]
res_df.reset_index(drop=True, inplace=True)
res_df.to_excel("SourceData-Fig3e.xlsx", sheet_name='Fig.3e')
#f.savefig("CT_radar_cell2loc.png", dpi=200, bbox_inches='tight')
```

```python
res = []
for method in ['stsc', 'spatialDWLS', 'RCTD', 'music', 'spotlight']:
    df, f = get_result_method(method, True)
    res.append(df)
    print(method)
    f.savefig(f"CT_radar_{method}.png", dpi=200, bbox_inches='tight')
res = pd.concat(res)
```

```python
res['grp'] = res[['method', 'reference']].apply(lambda x: '_'.join(x), axis=1)
res.to_csv("cell_type_performance_table.csv")
```
