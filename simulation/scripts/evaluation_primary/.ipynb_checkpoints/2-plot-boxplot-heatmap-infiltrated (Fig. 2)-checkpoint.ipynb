```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import numpy as np
import sys
import warnings
from scipy.stats import entropy
warnings.filterwarnings("ignore")
```

```python
###
methods = ['stereoscope', 'spatialDWLS', 'RCTD', 'music', 'spotlight']
cell_types = ['Cancer', 'Ductal', 'Stroma']
folder = '/houston_20t/alexw/ST/iSort/simulation_010222/s1/pure/'
truth = pd.read_csv(f"/houston_20t/alexw/ST/iSort/simulation_finer_resolution/data/pure/simulated_mixture_cell_ground_truth.csv", index_col=0)
truth = truth[cell_types]
truth = truth.divide(truth.sum(axis=1), axis=0)
```

```python
def read_process_res(fp, method, cell_types=cell_types):
    
    if method == 'stereoscope':
        sep = "\t"
    else:
        sep = ","
        
    df = pd.read_csv(fp, sep=sep, index_col=0)

    if method == 'spatialDWLS':
        df.set_index('cell_ID', drop=True, inplace=True)
        df.index.name = None
        
    df = df[cell_types]
    
    for i in range(df.shape[0]):
        if (df.iloc[i, :] == 0).all():
            df.iloc[i, :] = 0.1
            
    df = df.divide(df.sum(axis=1), axis=0)
    return df

def evaluate_performance(est, tru, 
                      cell_types=cell_types):
    """calculate cell type level Mean Absolute Percentage Error"""
    est = est.loc[tru.index, cell_types]
    tru = tru.loc[:, cell_types]
    mae = np.nanmean(np.absolute(est.values - tru.values))
    rs, kls = [], []
    for i in range(est.shape[0]):
        try:
            rs.append(pearsonr(est.values[i], tru.values[i])[0])
        except:
            rs.append(np.nan)
        kl = entropy(est.values[i]+ 0.001, tru.values[i]+ 0.001)
        
        kls.append(kl)
    ks_df = pd.DataFrame({'KL': kls}, index = tru.index)
    rs_df = pd.DataFrame({'r': rs}, index = tru.index)
    
    return rs_df, ks_df

def get_results(method, cell_types=cell_types):
    
    if method == 'stereoscope':
        fmt = "tsv"
    else:
        fmt = "csv"
        
    f = f"{folder}{method}_results"
    
    kls_dfs = []
    
    # internal
    df_int = read_process_res(f"{f}/estimated_proportions_ref_internal.{fmt}", method)
    rs_int, kl_int = evaluate_performance(df_int, truth)
    rs_int['reference'] = 'Internal'
    kl_int['reference'] = 'Internal'
    
    # external
    df_ext = read_process_res(f"{f}/estimated_proportions_ref_external.{fmt}", method)
    rs_ext, kl_ext = evaluate_performance(df_ext, truth)
    rs_ext['reference'] = 'External'
    kl_ext['reference'] = 'External'

    # mist
    df_mist = read_process_res(f"{f}/estimated_proportions_ref_MIST.{fmt}", method)
    rs_mist, kl_mist = evaluate_performance(df_mist, truth)
    rs_mist['reference'] = 'MIST'
    kl_mist['reference'] = 'MIST'
    
    rs_dfs = pd.concat([rs_int, rs_ext, rs_mist])
    kl_dfs = pd.concat([kl_int, kl_ext, kl_mist])
    rs_dfs['method'] = method
    kl_dfs['method'] = method
    return rs_dfs, kl_dfs
```

```python
methods = ['stereoscope', 'spatialDWLS', 'RCTD', 'music', 'spotlight']

r_dfs, kl_dfs = [], []

for method in methods:
    r_df, kl_df = get_results(method)
    r_dfs.append(r_df)
    kl_dfs.append(kl_df)
r_dfs = pd.concat(r_dfs)
kl_dfs = pd.concat(kl_dfs)
```

```python
f = plt.figure(figsize=(15, 6))
sns.boxplot(data=kl_dfs, x='method', y='KL', hue='reference', 
            hue_order = ['Internal', 'MIST', 'External'], order = methods,
              linewidth=0.5)
sns.despine();
plt.xticks(range(5), ['Stereoscope', 'spatialDWLS', 'RCTD', 'MuSiC', 'SPOTlight'],
          fontsize=15);
plt.xlabel("")
plt.tick_params(axis='y', labelsize=15)
plt.ylabel("KL Divergence", fontsize=18)
plt.legend(fontsize=15, bbox_to_anchor=(1.1, 0.8));
plt.savefig('Pure_KL_diver_boxplot.png', dpi=200, bbox_inches='tight')
```

```python
f = plt.figure(figsize=(15, 6))
sns.boxplot(data=r_dfs, x='method', y='r', hue='reference', 
            hue_order = ['Internal', 'MIST', 'External'], order = methods,
              linewidth=0.5);
sns.despine();

plt.xticks(range(5), ['Stereoscope', 'spatialDWLS', 'RCTD', 'MuSiC', 'SPOTlight'],
          fontsize=15);

plt.xlabel("")
plt.tick_params(axis='y', labelsize=15)
plt.ylabel("Spearman's r", fontsize=18)
plt.legend(fontsize=15,bbox_to_anchor=(1.01, 0.8));
plt.savefig('Spearman_r_boxplot.png', dpi=200, bbox_inches='tight')
```

```python
r_mean_df = pd.pivot_table(data=r_dfs, index='method', columns='reference', values='r', aggfunc='mean')
r_mean_df.index.name=None
r_mean_df.columns.name=None

r_mean_df = r_mean_df.loc[methods, ['Internal','MIST', 'External']]

f, ax = plt.subplots(1,1)

sns.heatmap(data=r_mean_df, annot=True, annot_kws={'weight':'bold', 'size':'16'}, cmap='Reds')
ax.set_yticklabels(['Stereoscope', 'spatialDWLS', 'RCTD', 'MuSiC', 'SPOTlight']);
ax.set_xticklabels(['Internal', 'MIST', 'External']);
ax.set_xlabel("")
ax.tick_params(axis='both', labelsize=13)
ax.set_ylabel("Spearman's corr. coef.", fontsize=15)
plt.savefig('Pure_Mean_Spearman_r_heatmap.png', dpi=200, bbox_inches='tight')
```

```python
kl_mean_df = pd.pivot_table(data=kl_dfs, index='method', columns='reference', values='KL', aggfunc='mean')
kl_mean_df.index.name=None
kl_mean_df.columns.name=None
kl_mean_df = kl_mean_df.loc[methods,
                          ['Internal', 'MIST','External']]

f, ax = plt.subplots(1,1)

sns.heatmap(data=kl_mean_df, annot=True, annot_kws={'weight':'bold', 'size':'16'}, cmap='Reds_r')
ax.set_yticklabels(['Stereoscope', 'spatialDWLS', 'RCTD', 'MuSiC', 'SPOTlight'],
          fontsize=15);
ax.set_xticklabels(['Internal', 'MIST', 'External'],
          fontsize=13);
ax.set_xlabel("")
ax.tick_params(axis='y', labelsize=13)
ax.set_ylabel("KL Divergence", fontsize=15)
plt.savefig('Pure_Mean_KL_diver_heatmap.png', dpi=200, bbox_inches='tight')
```
