```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.lines import Line2D
```

```python
pure_gt = pd.read_csv("../../s1/pure/simulated_spot_props_ground_truth.csv", index_col=0)
xs, ys = [], []

for i in pure_gt.index.tolist():
    xs.append(int(i.split("x")[0]))
    ys.append(int(i.split("x")[1]))

f = plt.figure(figsize=(6,6))
ax = plt.subplot(1,1,1)
ax.scatter(xs, ys, c=pure_gt.values)
ax.axis("off")

legend_elements = [Line2D([0], [0], marker='o', color='w', label='Cancer',
                          lw=0, markerfacecolor='red', markersize=12),
                   Line2D([0], [0], marker='o', color='w', label='Stroma',
                          lw=0, markerfacecolor='blue', markersize=12),
                   Line2D([0], [0], marker='o', color='w', label='Ductal',
                          lw=0, markerfacecolor='green', markersize=12)]
ax.legend(handles=legend_elements, bbox_to_anchor=(0.98, 0.7),
          fancybox=True, fontsize=14)
plt.savefig("pure_simulated_patterns.png", dpi=100, bbox_inches='tight')
```

```python
inf_gt = pd.read_csv("../../s1/cancer_inf_0.1/simulated_spot_counts_ground_truth.csv", index_col=0)
inf_gt = inf_gt.divide(inf_gt.sum(axis=1), axis=0)

xs, ys = [], []

for i in inf_gt.index.tolist():
    xs.append(int(i.split("x")[0]))
    ys.append(int(i.split("x")[1]))

f = plt.figure(figsize=(6,6))
ax = plt.subplot(1,1,1)
ax.scatter(xs, ys, c=inf_gt.values)
ax.axis("off")

legend_elements = [Line2D([0], [0], marker='o', color='w', label='Cancer',
                          lw=0, markerfacecolor='red', markersize=12),
                   Line2D([0], [0], marker='o', color='w', label='Stroma',
                          lw=0, markerfacecolor='blue', markersize=12),
                   Line2D([0], [0], marker='o', color='w', label='Ductal',
                          lw=0, markerfacecolor='green', markersize=12)]
ax.legend(handles=legend_elements, bbox_to_anchor=(0.98, 0.7),
          fancybox=True, fontsize=14)
plt.savefig("cancer_infilt_simulated_patterns.png", dpi=100, bbox_inches='tight')
```
