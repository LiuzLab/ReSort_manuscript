```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
## PURE
```

```python
meta = pd.read_csv("/houston_30t/alexw/ReSort_manuscript/simulation/reference/primary/ReSort/meta.csv", index_col=0)
meta.iloc[:4,:]
```

```python
meta['x'] = [int(i.split("x")[0]) for i in meta.index.tolist()]
meta['y'] = [int(i.split("x")[1]) for i in meta.index.tolist()]

f = plt.figure(figsize=(6,6))
cmap = {"Cancer": "red", "Stroma":"blue", "Ductal":"green"}
sns.scatterplot(x='x', y='y', data=meta, hue='bio_celltype', palette=cmap, legend=False)
plt.xticks([])
plt.yticks([])
plt.xlabel("")
plt.ylabel("")
plt.savefig("Supp-Fig1.png", dpi=100, bbox_inches='tight')
```
